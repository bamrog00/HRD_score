```python
import pandas as pd
from pathlib import Path
import glob
import os
import matplotlib.pyplot as plt
```

```python
directory = "../data/figures_pan_cancer/scores_correlation"

png_files = [f for f in os.listdir(directory) if f.endswith(".png")]

# Calculate the number of rows and columns needed
num_images = len(png_files)
num_cols = 3
num_rows = (num_images + num_cols - 1) // num_cols


# Create a subplot with multiple rows and columns
fig, axs = plt.subplots(nrows=num_rows, ncols=num_cols, figsize=(10, 10))

# Loop through each PNG file and plot it on a subplot
for i, png_file in enumerate(png_files):
    row = i // num_cols
    col = i % num_cols
    img = plt.imread(os.path.join(directory, png_file))
    axs[row, col].imshow(img)
    axs[row, col].axis('off')
    #axs[row, col].set_xlim(-0.8, 0.8)
    #axs[row, col].set_ylim(-0.8/aspect_ratio, 0.8/aspect_ratio)

# Delete any empty subplots
for i in range(num_rows):
    for j in range(num_cols):
        if i*num_cols+j >= num_images:
            fig.delaxes(axs[i, j])

# Adjust the spacing between subplots
plt.subplots_adjust(hspace=0.4, wspace=0.4)

# Save the plot as an image file
plt.savefig('output.png')

# Show the plot
plt.show()
```
